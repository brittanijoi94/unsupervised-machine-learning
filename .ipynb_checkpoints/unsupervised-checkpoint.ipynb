```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('crypto_data.csv')
df = df.loc[df['IsTrading'] == True]
df = df.dropna(how='any')
df = df.drop(['Unnamed: 0', 'IsTrading'], axis=1)
df = df.loc[df['TotalCoinsMined'] > 0]
df = df.drop(['CoinName'], axis=1)
#get dummies of algorithm and prooftype
df = pd.get_dummies(df, columns=['Algorithm', 'ProofType'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# df['TotalCoinsMined'] = scaler.fit_transform(df['TotalCoinsMined'].values.reshape(-1, 1))
df_scaled = scaler.fit_transform(df)
```

```python
#Dimensionality reduction with PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=0.99)
df_pca = pca.fit_transform(df)
```

```python
#Further reduce with t-SNE and plot the results
from sklearn.manifold import TSNE
tsne = TSNE(perplexity=40)
tsne_results = tsne.fit_transform(df_pca)
tsne_results = pd.DataFrame(tsne_results)
```

```python
 #Then create a scatter plot of the t-SNE output. Observe whether there are distinct clusters or not.
plt.scatter(tsne_results[0], tsne_results[1])
plt.show()
```

```python
inertia = []
from sklearn.cluster import KMeans
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(tsne_results)
    inertia.append(kmeans.inertia_)
elbow = pd.DataFrame(inertia, index=range(1, 11))

plt.plot(elbow.index, elbow.values)
plt.show()
```

3 is the best number of clusters.
